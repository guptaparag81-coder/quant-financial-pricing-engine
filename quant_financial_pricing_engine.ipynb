
# Quantitative Financial Pricing Engine

## Objective

Implement the Black-Scholes option pricing model for European call options.

## Technologies

## Mathematical Background

The Black-Scholes equation is a partial differential equation used to estimate the fair value of financial derivatives.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def black_scholes_call(S, K, T, r, sigma):

    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))

    d2 = d1 - sigma*np.sqrt(T)

    call_price = S*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)

    return call_price
```

```python
S = 100
K = 100
T = 1
r = 0.05
sigma = 0.2

price = black_scholes_call(S,K,T,r,sigma)

print("Call Option Price =", round(price,2))
```

```python
stock_prices = np.linspace(50,150,100)

option_prices = []

for S in stock_prices:
    option_prices.append(
        black_scholes_call(S,100,1,0.05,0.2)
    )

plt.figure(figsize=(8,5))
plt.plot(stock_prices, option_prices)

plt.xlabel("Stock Price")
plt.ylabel("Option Price")
plt.title("Black-Scholes Call Option Pricing")

plt.grid(True)

plt.show()
```
